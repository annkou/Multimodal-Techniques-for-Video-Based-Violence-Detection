```python
OUTPUT_VIDEO_FOLDER = "../data/videos"
```

```python
# Download the default config ("video" - actual video files)
video_urls_path = "../data/metadata/video_links_labels.csv"
video_urls = pd.read_csv(video_urls_path)
video_urls.head(3)
```

```python
def download_dataset(output_folder):
    for url in tqdm(video_urls["download_url"], desc="Downloading videos"):
        filename = os.path.basename(url)
        output_path = os.path.join(output_folder, filename)
        if os.path.exists(output_path):
            continue  # Skip if already downloaded
        try:
            with requests.get(url, stream=True, timeout=60) as r:
                r.raise_for_status()
                with open(output_path, "wb") as f:
                    for chunk in r.iter_content(chunk_size=8192):
                        if chunk:
                            f.write(chunk)
        except Exception as e:
            print(f"Failed to download {url}: {e}")


download_dataset(OUTPUT_VIDEO_FOLDER)
```
